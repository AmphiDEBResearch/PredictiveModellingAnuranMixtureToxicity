# Cross-validation of 2,4D TKTD model with an independent experiment 

The TKTD model of 2,4D effects in *Discoglossus* was originally calibrated on the UCLM larval growth data (time-resolved wet mass and number of tadpoles). <br>
In this analysis, we attempt to predict the 2,4d effects observed in the UGent dataset <br>
This qualifies as a cross-validation that addresses multiple aspects: 

- **Extrapolation across experiments**
- **Extrapolation across exposure patterns**: 
    - UCLM: constant exposure during the larval period. 
    - UGent: pulsed exposure at the end of larval period. 
- **Extrapolation across endpoints**: 
    - UCLM: 
        - time-resolved wet mass
        - time-resolved number of tadpoles (related to probability of reaching metamorphosis, not mortality)
    - UGent:
        - 10-day weight change after metamorphosis (effect measured after exposure ended)
